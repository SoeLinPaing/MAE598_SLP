### Problem 1

$$ 
\begin{aligned}
\quad & \min_{x_{1}, x_{2}} \quad f(x) = (x_{1} + 1)^2 +  (x_{2} - 2)^2 \\
\quad & s.t. \\
\quad & g_1 = x_{1} - 2 \le 0 \\
\quad & g_2 = x_{2} - 1 \le 0 \\
\quad & g_3 = - x_{1} \le 0 \\
\quad & g_4 = - x_{2} \le 0
\end{aligned}
$$

Find the optimum graphically. Determine directions of feasible descent at the corner
points of the feasible domain. Show the gradient directions of $f$ and $g_{i}s$ at these points.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle
```

```python
x1list = np.linspace(-2.0, 3.0, 100)
x2list = np.linspace(-2.0, 3.0, 100)
x1, x2 = np.meshgrid(x1list, x2list)
Z = (x1+1)**2 + (x2-2)**2

fig,ax=plt.subplots(1,1)
cp = ax.contourf(x1, x2, Z)

fig.colorbar(cp) # Add a colorbar to a plot
ax.add_patch(Rectangle((0, 0), 2, 1,
                    edgecolor='red',
                    facecolor='none',
                    lw=4))

plt.plot(-1, 2, marker="o", markersize=5, markeredgecolor="red", markerfacecolor="green")


ax.arrow(0, 1, -2/4, 2/4, width = 0.02)
ax.arrow(0, 1, 0.5, 0, width = 0.02)
ax.arrow(0, 1, 0, -0.5, width = 0.02)

ax.arrow(0, 0, -2/4, 4/4, width = 0.02)
ax.arrow(0, 0, 0.5, 0, width = 0.02)
ax.arrow(0, 0, 0, 0.5, width = 0.02)

ax.arrow(2, 1, -6/4, 2/4, width = 0.02)
ax.arrow(2, 1, 0, -0.5, width = 0.02)
ax.arrow(2, 1, -0.5, 0, width = 0.02)

ax.arrow(2, 0, -6/4, 4/4, width = 0.02)
ax.arrow(2, 0, 0, 0.5, width = 0.02)
ax.arrow(2, 0, -0.5, 0, width = 0.02)

ax.annotate('(-1,2)', xy = (-1, 2))

ax.annotate('-grad f', xy = (-0.3, 1.3))
ax.annotate('-grad g3', xy = (0.1, 1.1))
ax.annotate('-grad g2', xy = (0.1, 0.8))

ax.annotate('-grad f', xy = (-1, 0.5))
ax.annotate('-grad g3', xy = (0, -0.2))
ax.annotate('-grad g4', xy = (0, 0.3))

ax.annotate('-grad f', xy = (1.5, 1.2))
ax.annotate('-grad g1', xy = (1.2, 0.8))
ax.annotate('-grad g2', xy = (2, 0.8))

ax.annotate('-grad f', xy = (1., 0.5))
ax.annotate('-grad g1', xy = (1.5, -0.2))
ax.annotate('-grad g4', xy = (2, 0.3))

ax.set_title('$(x_1 + 1)^2 +  (x_2 - 2)^2$')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

As we can see in the figure, the optimal solution which satisfies the constraint is (0,1).

Lagragian can be formulated as follows:

$$ \mathcal{L} = (x_{1} + 1)^2 +  (x_{2} - 2)^2 + \mu_1 (x_{1} - 2) + \mu_2 (x_{2} - 1) + \mu_3 (- x_{1}) + \mu_4 (- x_{2}) $$

$${
\nabla_x \mathcal{L} 
=
\begin{bmatrix}
2 \ (x_{1} + 1) + \mu_1 - \mu_3  \\
2 \ (x_{2} - 2) + \mu_2 - \mu_4  \\
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
}$$



$${
\nabla_x f 
=
\begin{bmatrix}
2 \ (x_{1} + 1) \\
2 \ (x_{2} - 2)
\end{bmatrix}
}$$


$$ {
\nabla_x g_1 = \begin{bmatrix}
1\\
0
\end{bmatrix}
, 
\nabla_x g_2 = \begin{bmatrix}
0\\
1
\end{bmatrix}
, 
\nabla_x g_3 = \begin{bmatrix}
-1\\
0
\end{bmatrix}
, 
\nabla_x g_4 = \begin{bmatrix}
0\\
-1
\end{bmatrix}
} $$

---

For point $(x_1, x_2)=(0,0)$ , the constraints $g_3$ and $g_4$ are active which leads to $\mu_3 >0$, $\mu_4>0$, $\mu_1 =0$ and $\mu_2 =0$

$${
\nabla_x \mathcal{L} 
=
\begin{bmatrix}
2  - \mu_3  \\
-4  - \mu_4  \\
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
}$$

This gives $\mu_3 = 2$ and $\mu_4 = -4$ and as the $\mu_4$ is negative, KKT condition is not satisfied.

---

For point $(x_1, x_2)=(2,0)$ , the constraints $g_1$ and $g_4$ are active which leads to $\mu_1 >0$, $\mu_4>0$, $\mu_3 =0$ and $\mu_2 =0$

$${
\nabla_x \mathcal{L} 
=
\begin{bmatrix}
6 +\mu_1  \\
-4  - \mu_4  \\
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
}$$

This gives $\mu_1 = -6$ and $\mu_4 = -4$ and as the $\mu_1$ and $\mu_4$ are negative, KKT condition is not satisfied.

---

For point $(x_1, x_2)=(2,1)$ , the constraints $g_1$ and $g_2$ are active which leads to $\mu_1 >0$, $\mu_2>0$, $\mu_3 =0$ and $\mu_4 =0$

$${
\nabla_x \mathcal{L} 
=
\begin{bmatrix}
6 +\mu_1  \\
-2  + \mu_2  \\
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
}$$

This gives $\mu_1 = -6$ and $\mu_2 = 2$ and as the $\mu_1$ is negative, KKT condition is not satisfied.

---

For point $(x_1, x_2)=(0,1)$ , the constraints $g_2$ and $g_3$ are active which leads to $\mu_2 >0$, $\mu_3>0$, $\mu_1 =0$ and $\mu_4 =0$

$${
\nabla_x \mathcal{L} 
=
\begin{bmatrix}
2 -\mu_3  \\
-4  + \mu_2  \\
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
}$$

This gives $\mu_2 = 4$ and $\mu_3 = 2$ and as none of them are negative, KKT condition is satisfied.

Then, the solution is $(x_1, x_2)=(0,1)$

## Problem 2

$$ 
\begin{aligned}
\quad & \min \quad f(x) = -x_1 \\
\quad & s.t. \\
\quad & g_1 = x_{2} - (1-x_1)^3 \le 0 \\
\quad & g_2 = x_{2} \ge 0
\end{aligned}
$$

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x1 = np.linspace(-5, 5, 100)
f = -x1

plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True

ax = plt.subplot()
ax.plot(x1, f)

g = -(1-x1)**3
ax.plot(x1, g)

label = ax.set_xlabel('$x_1$')
label = ax.set_ylabel('$f$')
ax.set_ylim([-10, 10])
plt.grid()
plt.show()
```

Just by looking at the 2 dimensional plot setting $x_2 = 0$, we can see that there is no solution graphically.

Lagragian can be formulated as follows:

$$ \mathcal{L} = -x_1 + \mu_1 (x_{2} - (1-x_1)^3) + \mu_2 (- x_{2}) $$

$${
\nabla_x \mathcal{L} 
=
\begin{bmatrix}
-1 + 3\mu_1(1-x_1)^2   \\
\mu_1  - \mu_2  \\
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
}$$

$${
\nabla_x f 
=
\begin{bmatrix}
-1 \\
0
\end{bmatrix}
}$$

There are total 4 conditions to test:
- both $g_1$ and $g_2$ are active
- $g_1$ is active but $g_2$ is not
- $g_2$ is active but $g_1$ is not
- both $g_1$ and $g_2$ are inactive

When the constraints $g_1$ and $g_2$ are active, $\mu_1 > 0$ and $\mu_2 >0$

Since the point where all the constraints are met can be found using: $-x_1=x_{2} - (1-x_1)^3$ which is $x_1 =0.31767$ 

$${
\nabla_x \mathcal{L} 
=
\begin{bmatrix}
-1 + 3\mu_1(1-0.31767)^2   \\
\mu_1  - \mu_2  \\
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
}$$

## Problem 3

Objective function

$$ \min_{x_{1}, x_{2}, x_{3}} \quad f(x) = - x_{1} x_{2} - x_{2} x_{3} - x_{1} x_{3} $$

$$ s.t. \quad h(x) = x_{1} + x_{2} + x_{3} - 3 = 0 $$

Lagragian can be formulated as:

$$\mathcal{L} = -x_1x_2-x_2x_3-x_3x_1+\lambda(x_1+x_2+x_3-3)$$

$${
\nabla_x \mathcal{L} 
=
\begin{bmatrix}
-x_2-x_3+\lambda   \\
-x_1-x_3+\lambda  \\
-x_2-x_1+\lambda
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0 \\
0
\end{bmatrix}
}$$

$${
\nabla_\lambda \mathcal{L} 
= x_1+x_2+x_3-3=0
}$$

By solving the equations: $x_1 = x_2 = x_3 = 1$ and $\lambda = 2$

$${
\mathcal{L}_{xx} 
=
\begin{bmatrix}
0 & -1 & -1   \\
-1 & 0 & -1  \\
-1 & -1 & 0
\end{bmatrix}
}$$

To check second order sufficient condition:

$${
dx^T\mathcal{L}_{xx}dx 
=
\begin{bmatrix}
dx_1 & dx_2 & dx_3   \\
\end{bmatrix}
\begin{bmatrix}
0 & -1 & -1   \\
-1 & 0 & -1   \\
-1 & -1 & 0   
\end{bmatrix}
\begin{bmatrix}
dx_1   \\
dx_2   \\
dx_3   
\end{bmatrix}
}$$

$$= -2dx_1dx_2-2dx_1dx_3-2dx_2dx_3$$

$${
\frac{\partial h}{\partial x}dx 
= 0 \Rightarrow
\begin{bmatrix}
\frac{\partial h}{\partial x_1} & \frac{\partial h}{\partial x_2} & \frac{\partial h}{\partial x_3}   \\
\end{bmatrix}
\begin{bmatrix}
dx_1   \\
dx_2   \\
dx_3  
\end{bmatrix}
= 0
}$$

$${
\begin{bmatrix}
1 & 1 & 1   \\
\end{bmatrix}
\begin{bmatrix}
dx_1   \\
dx_2   \\
dx_3  
\end{bmatrix}
= 0
}$$

$$dx_1+dx_2+dx_3=0$$

$$dx_1 = -dx_2-dx_3$$

By substituting the above equation into $dx^T\mathcal{L}_{xx}dx$ 

$${dx^T\mathcal{L}_{xx}dx =
- 2 \ (- dx_2 - dx_3) \ dx_{2} - 2 \ dx_{2} \ dx_{3} - 2 \ (- dx_2 - dx_3) \ dx_{3} \\
= - 2 \ (- dx_2^2 - dx_{2} dx_{3} - dx_{2} dx_{3} - dx_{3}^2 d dx_{2} dx_{3}) \\
= 2 \ dx_2^2 + 2 \ dx_{3}^2 + 2 \ dx_{2} dx_{3} \\
 = 2 \ (dx_{2} + \frac{dx_{3}}{2})^2 + \frac{3 \ dx_{3}^2}{4} \ge 0
  }$$

if $dx_3 = 0$ and $dx_2 = 0$, then $dx_1 = 0$. Therefore, $(dx_2+\frac{1}{2}dx_3)^2+\frac{3}{4}dx_3^2 > 0$.
So, $dx^T\mathcal{L}_{xx}dx > 0$ when $dx$ is feasible.
