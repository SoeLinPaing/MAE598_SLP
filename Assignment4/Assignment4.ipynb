$$

## Problem 3

Objective function

$$ \min_{x_{1}, x_{2}, x_{3}} \quad f(x) = - x_{1} x_{2} - x_{2} x_{3} - x_{1} x_{3} $$

$$ s.t. \quad h(x) = x_{1} + x_{2} + x_{3} - 3 = 0 $$

Lagragian can be formulated as:

$$\mathcal{L} = -x_1x_2-x_2x_3-x_3x_1+\lambda(x_1+x_2+x_3-3)$$

$${
\nabla_x \mathcal{L} 
=
\begin{bmatrix}
-x_2-x_3+\lambda   \\
-x_1-x_3+\lambda  \\
-x_2-x_1+\lambda
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0 \\
0
\end{bmatrix}
}$$

$${
\nabla_\lambda \mathcal{L} 
= x_1+x_2+x_3-3=0
}$$

By solving the equations: $x_1 = x_2 = x_3 = 1$ and $\lambda = 2$

$${
\mathcal{L}_{xx} 
=
\begin{bmatrix}
0 & -1 & -1   \\
-1 & 0 & -1  \\
-1 & -1 & 0
\end{bmatrix}
}$$

To check second order sufficient condition:

$${
dx^T\mathcal{L}_{xx}dx 
=
\begin{bmatrix}
dx_1 & dx_2 & dx_3   \\
\end{bmatrix}
\begin{bmatrix}
0 & -1 & -1   \\
-1 & 0 & -1   \\
-1 & -1 & 0   
\end{bmatrix}
\begin{bmatrix}
dx_1   \\
dx_2   \\
dx_3   
\end{bmatrix}
}$$

$$= -2dx_1dx_2-2dx_1dx_3-2dx_2dx_3$$

$${
\frac{\partial h}{\partial x}dx 
= 0 \Rightarrow
\begin{bmatrix}
\frac{\partial h}{\partial x_1} & \frac{\partial h}{\partial x_2} & \frac{\partial h}{\partial x_3}   \\
\end{bmatrix}
\begin{bmatrix}
dx_1   \\
dx_2   \\
dx_3  
\end{bmatrix}
= 0
}$$

$${
\begin{bmatrix}
1 & 1 & 1   \\
\end{bmatrix}
\begin{bmatrix}
dx_1   \\
dx_2   \\
dx_3  
\end{bmatrix}
= 0
}$$

$$dx_1+dx_2+dx_3=0$$

$$dx_1 = -dx_2-dx_3$$

By substituting the above equation into $dx^T\mathcal{L}_{xx}dx$ 

$${dx^T\mathcal{L}_{xx}dx =
- 2 \ (- dx_2 - dx_3) \ dx_{2} - 2 \ dx_{2} \ dx_{3} - 2 \ (- dx_2 - dx_3) \ dx_{3} \\
= - 2 \ (- dx_2^2 - dx_{2} dx_{3} - dx_{2} dx_{3} - dx_{3}^2 d dx_{2} dx_{3}) \\
= 2 \ dx_2^2 + 2 \ dx_{3}^2 + 2 \ dx_{2} dx_{3} \\
 = 2 \ (dx_{2} + \frac{dx_{3}}{2})^2 + \frac{3 \ dx_{3}^2}{4} \ge 0
  }$$

if $dx_3 = 0$ and $dx_2 = 0$, then $dx_1 = 0$. Therefore, $(dx_2+\frac{1}{2}dx_3)^2+\frac{3}{4}dx_3^2 > 0$.
So, $dx^T\mathcal{L}_{xx}dx > 0$ when $dx$ is feasible.

## Problem 4

$$ \min_{x_{1}, x_{2}, x_{3}} \quad f(x) = x_{1}^2 + x_{2}^2 + x_{3}^2 $$

$$ 
\begin{aligned}
\quad & s.t. \\
\quad & h_1(x) = \frac{x_{1}^2}{4} + \frac{x_{2}^2}{5} + \frac{x_{3}^2}{25} - 1 = 0
\\
\quad & h_2(x) = x_{1} + x_{2} - x_{3} = 0
\end{aligned}
$$


Define objective functions and constraints.

```python
def cost(x1, x2, x3):
    return x1**2 + x2**2 + x3**2

def constraints(x1, x2, x3):
    h1 = (x1**2)/4 + (x2**2)/5 + (x3**2)/25 -1
    h2 = x1+x2-x3
    return h1, h2
```

We choose $x_1$ as decision variable and $x_2$ and $x_3$ as state variables

Calculate reduced gradient.

```python
#d = x1, s = x2, x3
def reduced_gradient(x1, x2, x3):
    dfds = np.vstack((2 * x2, 2 * x3))
    dfdeld = np.array([2 * x1])
    dhds = np.vstack((np.hstack(((2/5) * x2, (2/25) * x3)), np.array([[1, -1]], dtype=float)))
    dhdd = np.vstack(((1 / 2) * x1, 1.))
    dhds_inv = np.linalg.inv(dhds)
    dfdd = dfdeld - np.matmul(dfds.T, np.matmul(dhds_inv, dhdd))
    return dfds, dfdeld, dhds, dhdd, dhds_inv, dfdd
```

Line Search

```python
def line_search(x1, x2, x3):
    alpha = 1.0
    b = 0.5
    t = 0.3
    
    #calculate Phi
    def phi(x1, x2, x3, alpha):
        dfds, dfdeld, dhds, dhdd, dhds_inv,dfdd = reduced_gradient(x1, x2, x3)
        return cost(x1, x2, x3) - alpha * t * np.matmul(dfdd, dfdd.T)  

    #calculate f
    def f(x1, x2, x3, alpha):
        x1 = x1 - alpha * dfdd.flatten()
        ds = np.matmul(np.matmul(dhds_inv, dhdd), dfdd.T).flatten()
        x2 = x2 + alpha * ds[0]
        x3 = x3 + alpha * ds[1]
        return cost(x1, x2, x3)

    phiv = phi(x1, x2, x3, alpha)
    fv = f(x1, x2, x3, alpha)
    
    while phiv < fv:
        alpha = 0.5 * alpha
        phiv = phi(x1, x2, x3, alpha)
        fv = f(x1, x2, x3, alpha)
    return alpha
```

```python
def solvefun(x1, x2, x3):
    criterion = 1e-3
    h1, h2 = constraints(x1, x2, x3)
    h = np.vstack((h1, h2))
    h_norm = np.linalg.norm(h)

    while h_norm > criterion:
        dfds, dfdeld, dhds, dhdd, dhds_inv, dfdd = reduced_gradient(x1, x2, x3)
        ds = np.matmul(dhds_inv, h)
        x2 = x2 - ds[0]
        x3 = x3 - ds[1]
        h1, h2 = constraints(x1, x2, x3)
        h = np.vstack((h1, h2))
        h_norm = np.linalg.norm(h)
    return x1, x2, x3, h_norm
```

```python
eps = 1e-3 
x1 = np.array([1.0])
x2 = np.array([1.0])
x3 = np.array([1.0])

x1, x2, x3, h_norm = solvefun(x1, x2, x3)
dfds, dfdeld, dhds, dhdd, dhds_inv, dfdd = reduced_gradient(x1, x2, x3)
dfdd_norm = np.linalg.norm(dfdd)

while dfdd_norm >= eps:
    alpha = line_search(x1, x2, x3)
    dfds, dfdeld, dhds, dhdd, dhds_inv, dfdd = reduced_gradient(x1, x2, x3)
    x1 = x1 - alpha * dfdd.flatten()
    dfds, dfdeld, dhds, dhdd, dhds_inv, dfdd = reduced_gradient(x1, x2, x3)
    ds = alpha * np.matmul(np.matmul(dhds_inv, dhdd), dfdd.T).flatten()
    x2 = x2 + ds[0]
    x3 = x3 + ds[1]
    x1, x2, x3, h_norm = solvefun(x1, x2, x3)
    dfds, dfdeld, dhds, dhdd, dhds_inv, dfdd = reduced_gradient(x1, x2, x3)
    dfdd_norm = np.linalg.norm(dfdd)

print('x1={}, x2={}, x3={})'.format(x1, x2, x3))
```

## Problem 5

There are total N sites for the garbage truck to visit and $c_{ij}$ is the cost of the travel from site $i$ to $j$ and $\infty$ if there is no path between them.

Let $x_{ij}$ be the integer variable which represents 1 when the route goes from site $i$ to $j$ and 0 otherwise.

Since site 0 is the truck station, the problem can be formulated as:

$${
\text{min} \sum_{i=0}^{N}\sum_{j=0}^{N} c_{ij}x_{ij}
}$$

Since we don't want to go back to already visited sites, the constraints can be written as follows.

To constraint the route to go exactly to other site different from the one we just left:

$${
\sum_{i=0}^{N} x_{ij} = 1, \quad \text{for}\quad j = 0,1,2,...N \quad \text{and} \quad i\neq j
}$$

To constraint the route to go exactly to other site different from the one we just arrives:

$${
\sum_{k=0}^{N} x_{jk} = 1, \quad \text{for}\quad j = 0,1,2,...N \quad \text{and} \quad k\neq j
}$$

Then the complete problem formulation is:

$${
\text{min} \sum_{i=0}^{N}\sum_{j=0}^{N} c_{ij}x_{ij}
}$$

Subject to:

$${
\sum_{i=0}^{N} x_{ij} = 1, \quad \text{for}\quad j = 0,1,2,...N \quad \text{and} \quad i\neq j
}$$

$${
\sum_{k=0}^{N} x_{jk} = 1, \quad \text{for}\quad j = 0,1,2,...N \quad \text{and} \quad k\neq j
}$$
